```python
import requests
from requests_file import FileAdapter
import pandas as pd
from bs4 import BeautifulSoup
```

```python
data=pd.read_csv(r'D:\ai\Downloads\Data\paintings\archive\wclasses.csv')
```

```python
new_data=data[data['genre']==139].reset_index(drop=True)
new_data['search']=new_data['file'].str.split('/').str[1]
new_data['search']=new_data['search'].str.replace('_','/')
new_data['search']=new_data['search'].str.rsplit('.').str[0]
new_data['search']=new_data['search'].str.rsplit('(').str[0]
new_data
```

```python
new_data.to_csv(r'D:\ai\Downloads\Data\paintings\archive\new_csv.csv',index=False)
```

```python
url = 'https://www.wikiart.org/en/'
step = 1000
end = len(new_data)
for j in range(0,end,step):
    new_data=pd.read_csv(r'D:\ai\Downloads\Data\paintings\archive\new_csv.csv')
    for i in range(j,j+step):
        try:
            response = requests.get(url+new_data.loc[i,'search'])
            soup = BeautifulSoup(response.text, 'html.parser')
            genre = soup.select_one('li.dictionary-values > span > a> span').text
            new_data.loc[i,'genre']=genre
        except:
            print(i,'번째 실패')
    new_data.to_csv(r'D:\ai\Downloads\Data\paintings\archive\new_csv.csv',index=False)
    print(j+step,'완료')
```

```python
new_data=new_data[~new_data['genre'].astype(str).str.isdigit()]
```

```python
new_data.to_csv(r'D:\ai\Downloads\Data\paintings\archive\new_csv.csv',index=False)
```
