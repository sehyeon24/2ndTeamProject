# csv파일 가공

```python
data = pd.read_csv('../file/wclasses.csv')
data.head()
```

```python
# 그림명
data['new_filename'] = data['file'].str.split('/').str[1].str.split('_').str[1].str.rsplit('.', 1).str[0]
data.tail()
```

```python
# 화가
data['new_artist'] = data['file'].str.split('/').str[1].str.split('_').str[0]
data.head()
```

```python
# 장르
genre_mapping = {
    129: 'abstract',
    130: 'cityscape',
    131: 'genre painting',
    132: 'illustration',
    133: 'landscape',
    134: 'nude painting',
    135: 'portrait',
    136: 'religious painting',
    137: 'sketch and study',
    138: 'still life',
    139: '미분류'
}

data['new_genre'] = data['genre'].map(genre_mapping)
data.head()
```

```python
# 스타일
style_mapping = {
    140: 'Abstract_Expressionism',
    141: 'Action_painting',
    142: 'Analytical_Cubism',
    143: 'Art_Nouveau_Modern',
    144: 'Baroque',
    145: 'Color_Field_Painting',
    146: 'Contemporary_Realism',
    147: 'Cubism',
    148: 'Early_Renaissance',
    149: 'Expressionism',
    150: 'Fauvism',
    151: 'High_Renaissance',
    152: 'Impressionism',
    153: 'Mannerism_Late_Renaissance',
    154: 'Minimalism',
    155: 'Naive_Art_Primitivism',
    156: 'New_Realism',
    157: 'Northern_Renaissance',
    158: 'Pointillism',
    159: 'Pop_Art',
    160: 'Post_Impressionism',
    161: 'Realism',
    162: 'Rococo',
    163: 'Romanticism',
    164: 'Symbolism',
    165: 'Synthetic_Cubism',
    166: 'Ukiyo_e'
}

data['new_style'] = data['style'].map(style_mapping)
data.head()
```

```python
# 파일경로
import os
folder_path = "D:/imsehyeon/source/2ndTeamProject/file/"
data['new_filepath'] = data['file'].apply(lambda x: os.path.join(folder_path, x))
data.head()
```

```python
# style 139
search_genre = pd.read_csv('new_csv.csv')
search_genre
```

```python
data['new_genre'] = data.apply(
    lambda row: search_genre.loc[search_genre['file'] == row['file'], 'genre'].values[0] 
    if row['new_genre'] == '미분류' and not search_genre.loc[search_genre['file'] == row['file'], 'genre'].empty 
    else row['new_genre'], axis=1
)
data.head()
```

```python
data[data['new_genre'] == '미분류']
```

```python
data = data[data['new_genre'] != '미분류']
```

```python
data = data.drop(['artist', 'genre', 'style', 'new_filepath'], axis=1)
data = data.rename({'new_filename':'title', 'new_artist':'artist', 'new_genre':'genre', 'new_style':'style'}, axis=1)
```

```python
filepath = '../data/'
data = data[data['file'].apply(lambda x: os.path.exists(os.path.join(filepath, x)))].reset_index(drop=True)
```

```python
data.to_csv('painting.csv', index=False)
```
